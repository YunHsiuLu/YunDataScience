```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
mu=0
sigma=1
x=np.linspace(-4+mu, 4+mu, 800)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(%d,%d)' % (mu, sigma))
plt.show()
```

```python
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(%d,%d)' % (mu, sigma))
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print("The probability when x <= 1: %.2f" % (cumsum_probs[500]*100) + "%")
# cumsum_probs[500] => x <= 1
```

```python
print("The probability when x >= 1 and x <= -1: %.2f" % ((1 - cumsum_probs[500] + cumsum_probs[300])*100) + "%")
# 1 - cumsum_probs[500] => x >= 1
# cumsum_probs[300]     => x <= -1
```

```python
mu1=2
sigma1=4
x1=np.linspace(-20+mu1, 20+mu1, 4000)
# 2.產生 pdf圖
y1 = stats.norm.pdf(x1,mu1, sigma1)
plt.plot(x1, y1, "r--", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(%d,%d)' % (mu1, sigma1))
plt.show()
cumsum_probs =  stats.norm.cdf(x1,mu1, sigma1)
print("The probability when x <= 3: %.2f" % (cumsum_probs[2300]*100) + "%")
```
