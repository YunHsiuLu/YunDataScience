```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from scipy import stats
```

```python
def plotFig(sample_mean, x0, y0, x1, y1):
    fig, ax = plt.subplots()
    ix0, iy0, ix1, iy1 = np.array([]), np.array([]), np.array([]), np.array([])
    ix11, iy11, ix12, iy12 = np.array([]), np.array([]), np.array([]), np.array([])
    if sample_mean == 1:
        ix0 = x0[705:]
        iy0 = y0[705:]
        ix1 = x1[0:496]
        iy1 = y1[0:496]
        verts0 = [(sample_mean + 0.1, 0)] + list(zip(ix0, iy0)) + [(4, 0)]
        poly0 = Polygon(verts0, facecolor = 'purple', alpha=0.3)
        ax.add_patch(poly0)
        verts1 = [(-2, 0)] + list(zip(ix1, iy1)) + [(sample_mean - 0.1, 0)]
        poly1 = Polygon(verts1, facecolor = 'red', alpha=0.3)
        ax.add_patch(poly1)
    elif sample_mean > 1:
        ix0 = x0[int((sample_mean - (-6))*100 + 5):]
        iy0 = y0[int((sample_mean - (-6))*100 + 5):]
        ix11 = x1[0:501]
        iy11 = y1[0:501]
        ix12 = x0[701:int((sample_mean - (-6))*100 - 5)]
        iy12 = y0[701:int((sample_mean - (-6))*100 - 5)]
        verts0 = [(sample_mean + 0.1, 0)] + list(zip(ix0, iy0)) + [(4, 0)]
        poly0 = Polygon(verts0, facecolor = 'purple', alpha=0.3)
        ax.add_patch(poly0)
        verts1 = [(-2, 0)] + list(zip(ix11, iy11)) + list(zip(ix12, iy12)) + [(sample_mean - 0.1, 0)]
        poly1 = Polygon(verts1, facecolor = 'red', alpha=0.3)
        ax.add_patch(poly1)
    else:
        ix1 = x1[0:int((sample_mean - (-4))*100 + 1 - 5)]
        iy1 = y1[0:int((sample_mean - (-4))*100 + 1 - 5)]
        ix11 = x0[700:]
        iy11 = y0[700:]
        ix12 = x1[int((sample_mean - (-4))*100 + 1 + 5):501]
        iy12 = y1[int((sample_mean - (-4))*100 + 1 + 5):501]
        verts0 = [(sample_mean + 0.1, 0)] + list(zip(ix12, iy12)) + list(zip(ix11, iy11)) + [(4, 0)]
        poly0 = Polygon(verts0, facecolor = 'purple', alpha=0.3)
        ax.add_patch(poly0)
        verts1 = [(-2, 0)] + list(zip(ix1, iy1)) + [(sample_mean - 0.1, 0)]
        poly1 = Polygon(verts1, facecolor = 'red', alpha=0.3)
        ax.add_patch(poly1)
    plt.plot(x0, y0, "b-", linewidth=2, label="H0")
    plt.plot(x1, y1, "g-", linewidth=2, label="H1")
    plt.title('sample mean = %.2f' % (sample_mean))
    plt.legend(loc="upper right")
    plt.show()
```

```python
# x bar > 1; reject H0
sample_mean = 1
mu1 = 0
sigma = 1
mu2 = 2
x0 = np.linspace(-6 + mu1, 6 + mu1, 1200)
y0 = stats.norm.pdf(x0, mu1, sigma)
x1 = np.linspace(-6 + mu2, 6 + mu2, 1200)
y1 = stats.norm.pdf(x1, mu2, sigma)

plotFig(sample_mean, x0, y0, x1, y1)
```

```python
# x bar > 2; reject H0
xbar = 2
plotFig(xbar, x0, y0, x1, y1)
```

$\alpha$ will be smaller; $\beta$ will be larger

```python
# x bar < 0.2; reject H0
xbar = 0.2
plotFig(xbar, x0, y0, x1, y1)
```

$\alpha$ will be larger; $\beta$ will be smaller
