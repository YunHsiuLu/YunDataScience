```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1.   畫出 cos 圖檔，並儲存
f1 = plt.figure()
x = np.arange(0, 360)
y = np.cos(x * np.pi / 180)
plt.plot(x, y)
#plt.show()
f1.savefig('D18_exercise.png')
```

```python
N = 1024
a = np.random.normal(0, 1, N)
b = np.random.normal(0, 1, N)
colors = np.random.rand(N)
Sin = b / np.sqrt(a**2 + b**2)
Cos = a / np.sqrt(a**2 + b**2)
for i in range(N):
    if Sin[i] > 0:
        if Cos[i] < 0:
            colors[i] = (-Cos[i]) * 0.25 + 0.75
        elif Cos[i] > 0:
            colors[i] = Sin[i] * 0.25 + 0.5
        else:
            colors[i] = 0
    elif Sin[i] < 0:
        if Cos[i] > 0:
            colors[i] = Cos[i] * 0.25 + 0.25
        elif Cos[i] < 0:
            colors[i] = (-Sin[i]) * 0.25 + 0.0
        else:
            colors[i] = 0
    else:
        colors[i] = 0
# colors = np.actan2(b, a)
plt.scatter(a, b, s = 100, c=colors, alpha = 0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```
